## Nos Données (Our Data)  
_Dans ce cahier, nos variables proviennent d'un jeu de données de composants Icebreaker. Nous allons maintenant lire ce jeu de données sous forme de DataFrame dans pandas._  

<span style="color:gray">(In this notebook, our variables are sourced from a dataset of icebreaker components. We'll read in the dataset now as a DataFrame in pandas.</span>

```python
# Importez pandas (Import pandas)
import pandas as pd

# Importez random (Import random)
# Définissez la graine aléatoire à 42 pour permettre aux autres 
# de reproduire nos choix aléatoires. 
# (Define random seed as 42 so others can recreate our random choices)
import random
random.seed(42)

# Lisez notre jeu de données dans pandas en tant que DataFrame
# (Read our dataset into pandas as a DataFrame)
df = pd.read_csv('./data/breakfast_hour.csv')
df.head()
```

## Étape 1 : Créer un tableau qui stocke nos trois types de questions brise-glace.

_Tout d'abord, on va créer un tableau qui inclut nos trois types de questions :_

- _des questions avec des options 'nulles'_
- _des questions avec des options 'bonnes'_
- _des questions à réponses 'ouvertes'_

_Ensuite, nous assignerons ce tableau à la variable 'questions_fr', pour indiquer que ces questions doivent être en français._  


#### <span style="color:gray">Step 1: Create an array that stores our three icebreaker question types.</span> 

<span style="color:gray">First, we will create an array that includes our three question types:</span>
- <span style="color:gray">questions with 'bad' options</span>
- <span style="color:gray">questions with 'good' options</span>
- <span style="color:gray">'open' ended questions</span>

<span style="color:gray">Then we'll assign this array to variable 'questions_fr', to denote that these questions are to be in French.</span>

```python
# Créer un tableau qui stocke nos trois types de questions
# Sauvegarder le tableau dans la variable 'questions_fr'

# (Create array that stores our three question types)
# (Save array to variable 'questions_fr')
questions_fr = ['nulles', 'bonnes', 'ouvertes']
```

## Étape 2 : Demander au générateur de choisir au hasard dans le tableau.  
Je vais créer une variable 'brise_glace' qui choisira parmi les éléments de 'questions_fr' de manière aléatoire.
- Si 'brise_glace' choisit 'nulles' dans 'questions_fr', elle doit afficher une question avec deux mauvaises options.
- Si 'brise_glace' choisit 'bonnes', elle doit afficher une question qui propose deux bonnes options.
- Sinon, elle doit afficher une question à réponse ouverte.

<span style="color:gray">**Step 2: Tell the generator to pick from the array at random.**</span>  
<span style="color:gray">I will create a variable 'brise_glace' that will pick from the elements in 'questions_fr' at so-called random.</span>  
- <span style="color:gray">If 'brise_glace' picks 'nulles' from 'questions_fr', it must print a question with two bad options.</span>
- <span style="color:gray">If 'brise_glace' picks 'bonnes', it must print a question that offers two good options.</span>
- <span style="color:gray">Otherwise, it must print an open-ended question.</span>

```python
# Créer 'brise_glace' pour choisir un type de question au hasard dans 'questions_fr'
# Si cela choisit 'nulle', afficher une question avec une option négative
# Si 'bonne', afficher une question avec une option positive
# Sinon, afficher une question à réponse ouverte

# (Create 'brise_glace' to pick a question type from 'questions_fr' at random)
# (If it picks 'nulle', print a negative-option question)
# (If 'bonne', print positive-option question)
# (Else print an open-ended question)

brise_glace = (random.choice(questions_fr))
if brise_glace == 'nulles':
    # Select two different bad options by sampling without replacement
    options = random.sample(df['mauvaises_préfs'].tolist(), 2)
    print(f'Vous préféreriez {options[0]}, ou {options[1]}?')
elif brise_glace == 'bonnes':
    # Select two different bad options by sampling without replacement
    options = random.sample(df['bonnes_préfs'].tolist(), 2)
    print(f'Vous préféreriez {options[0]}, ou {options[1]}?')
else: print(f"{random.choice(df['ouvertes'])}")
```
